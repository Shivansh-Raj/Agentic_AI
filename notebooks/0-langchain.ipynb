# Introduction to LangChain
This notebook covers the basics of LangChain, a framework for developing applications powered by language models.

## 1. Setup API Key
First, let's set up the OpenAI API key. You will need your own key from `platform.openai.com`.

```python
import os
import getpass
from dotenv import load_dotenv

load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

## 2. Basic LLM Call
Let's initialize an OpenAI language model and make a simple prediction.

```python
from langchain_groq import ChatGroq

# Initialize the model
llm = ChatGroq(model="llama-3.1-8b-instant")

# Make a prediction
response = llm.invoke("What is the capital of Italy?")
print(response.content)
```

## 3. Prompt Templates
Prompt templates make it easy to dynamically generate prompts with user input.

```python
from langchain_core.prompts import PromptTemplate

template = "What is a good name for a company that makes {product}?"
prompt = PromptTemplate.from_template(template)

# Format the prompt to see what it looks like before sending to the LLM
formatted_prompt = prompt.format(product="colorful socks")
print("Formatted Prompt:\n", formatted_prompt)
```

## 4. Chains with LCEL (LangChain Expression Language)
Chains allow us to combine multiple components together. The most common is chaining a PromptTemplate with an LLM using the pipe `|` operator.

```python
from langchain_core.output_parsers import StrOutputParser

# Create a simple chain: Prompt -> LLM -> Output Parser (to extract just the string content)
chain = prompt | llm | StrOutputParser()

# Run the chain
result = chain.invoke({"product": "colorful socks"})
print("Result:\n", result)
```
