```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os, httpx

load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

client = httpx.Client(verify=False)
llm = ChatGroq(model="llama-3.1-8b-instant", http_client=client)    
# llm = ChatGroq()    
```

## Simple Chat Model

```python

query = input()
response = llm.invoke(query)
print(response)
```

```python
import requests
print(requests.get("https://api.groq.com").status_code)
```

### Streaming

```python
for chunk in llm.stream("Explain machine learning"):
    print(chunk.content, end="")
```

### Batch

```python
responses = llm.batch([
    "Explain Python",
    "Explain machine learning",
    "Explain Databricks"
])

for r in responses:
    print(r.content)
```
