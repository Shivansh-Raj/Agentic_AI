## MESSAGES
🔹 One-line definition 

Messages = structured communication units that represent a turn in a dialogue system

```python
import os
from langchain_groq import ChatGroq

groq = ChatGroq(model_name="llama-3.1-8b-instant")
```

### Example 1 : 

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage

messages = [
    SystemMessage(content= "You are a Gen Z drunk teenager"),
    HumanMessage(content="What is the capital of France?"),
    AIMessage(content="New Delhi, obviously"),
    HumanMessage(content="Ohh, yes. France's Capital recently got changed.")
]
response = groq.invoke(messages)
response.content
# messages
```

### Messages MetaData
#### Example 2: Using metadata (for system, not LLM)

```python
messages = [
    SystemMessage(
        content="You are a drunk Gen Z professor",
        metadata={"persona": "formal"}
    ),
    
    HumanMessage(
        content="What is the capital of France?",
        metadata={"user_id": 101}
    ),
    
    AIMessage(
        content="New Delhi",
        metadata={"confidence": 1}
    ),
    
    HumanMessage(
        content="I think you are right.",
        metadata={"priority": "high"}
    )
]

response = groq.invoke(messages)
print(response.content)
```

#### Example 3: Metadata-based routing

```python

msg = HumanMessage(
    content="What is 6 * 3?",
    metadata={"mode": "lenient"}
)

if msg.metadata["mode"] == "strict":
    system_prompt = "Answer only with the final number. No explanation."
else:
    system_prompt = "Explain your answer."

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": msg.content}
]

response = groq.invoke(messages)
print(response.content)
```

### ToolMessage

```python
messages = [
    HumanMessage(content="What is 6 * 3?")
]

# Step 1: LLM decides to use tool
ai_msg = AIMessage(
    content="",
    tool_calls=[{
        "id": "call_1",
        "name": "alculator",
        "args": {"expression": "6 * 3"}
    }]
)

messages.append(ai_msg)

# Step 2: Tool executes
tool_result = "19"

# Step 3: Send result back as ToolMessage
messages.append(
    ToolMessage(
        content=tool_result,
        tool_call_id="call_1"
    )
)

# Step 4: LLM gives final answer
response = groq.invoke(messages)
print(response.content)
```
