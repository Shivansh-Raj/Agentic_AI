## Agents with Memory

```python
import os
from langchain_groq import ChatGroq

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
llm = ChatGroq(model_name="llama-3.1-8b-instant")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
agent = create_agent(
    model = llm,
    checkpointer = InMemorySaver()
)
```

### Creating a thread
Thread = identifier for a conversation session

```python
config = {"configurable":{"thread_id":"test1"}}
```

```python
from langchain.messages import HumanMessage, SystemMessage

response = agent.invoke(
    {"messages": [
        SystemMessage(content="Hello you are a genz professor that talks in a genz way"),
        HumanMessage(content="Yo wassup, homie I'm Shivansh.")
    ]},
    config=config
)
response
```

```python
agent.invoke({
    "messages":[HumanMessage("How old was I?")]
}, config=config)
```
